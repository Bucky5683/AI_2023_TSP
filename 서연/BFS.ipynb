```python
from collections import deque
import random
import math
import pandas as pd
import time
import csv
from scipy.spatial import distance
```

```python
csv_file = "../Project#1/2023_AI_TSP.csv"
solution_file = "solution.csv"
GA_Interation = 10000
running_time = 1000 #1ì¡°
start_time = time.time()
```

```python
#ë°ì´í° ì¢íë¥¼ ë¦¬ì¤í¸ë¡ ë³ê²½
data = pd.read_csv(csv_file, header=None, names=['x', 'y'])

list_data = data.values.tolist() #ì¢í ë¦¬ì¤í¸

len_data = len(list_data) #íì ê°ì

print(list_data)
print(len_data)
```

```python
# ê±°ë¦¬ ê³ì° í¨ì
# ê±°ë¦¬ : node 1ê³¼ node2ì ì§ì ê±°ë¦¬(í¼íê³ ë¼ì¤ ì ë¦¬)
# node = list_dataì idex
def distance_between_node(node1, node2):
    distance_x = list_data[node1][0] - list_data[node2][0]
    distance_y = list_data[node1][1] - list_data[node2][1]

    distace_pow = math.pow(distance_x, 2) + math.pow(distance_y, 2)

    return math.sqrt(distace_pow)

print(list_data[0]+list_data[1])
distance_between_node(0, 1)
```

```python
# ê²½ë¡ íê° í¨ì -> ëª¨ë  ë¸ë ìííë©° ê±°ë¦¬ ê³ì°
def evaluate_path(path):
    dist = 0
    for i in range(0, len(path)):
        node1 = path[i]
        node2 = path[(i+1)%len(path)]

        dist += distance_between_node(node1, node2)
    return dist

evaluate_path([0, 1, 2])
```

```python
# ë¬´ììë¡ í´ë¥¼ ë§ë¦
def make_path_list( ):
    path_list = []
    for i in range(running_time):
        #print('path running_time : ', i)
        path = list(range(1, len_data)) #ëª¨ë  ë¸ëì indexë¥¼ ë¦¬ì¤í¸ë¡ ìì±
        random.shuffle(path) #ë¸ë index ììë¥¼ ìì
        if path not in path_list:
            path.insert(0, 0)
            path_list.append(path)
    return path_list
```

```python
# ëª¨ë  ê²½ë¡ë¥¼ ìííì¬ ìµì í´ ì°¾ê¸°
def find_best_path(run_count):
    best_path = []
    dist = float('inf')
    for i in range(run_count):
        print('run_count : ', i)
        pathlist = make_path_list()
        for path in pathlist:
            candidate_dist = evaluate_path(path)
            if candidate_dist < dist:
                best_path = path
                dist = candidate_dist
    f = open(solution_file, 'w')
    for i in best_path:
        f.write(str(i) + '\n')
    f.close()
    return dist
'''
for path in path_list:
    dist = evaluate_path(path)
    if dist < best_dist:
        best_path = path
        best_dist = dist

    print("ìµì  ê±°ë¦¬: ", best_dist)
'''
```
